```python
x=df.iloc[:,[3,4]].values
x
```

```python
#Visualizing_Dendogram
import scipy.cluster.hierarchy as h
plt.figure(figsize=(16,8))
h.dendrogram(h.linkage(x,method='ward'))
plt.show()
```

```python
#Agglomerative_Clustering
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
y_h=hc.fit_predict(x)
y_h
```

```python
#visualizing_the_clusters_using_scatterplot
plt.figure(figsize=(15,7))
plt.scatter(x[y_h==0,0],x[y_h==0,1],color='blue',label='cluster1',s=30)
plt.scatter(x[y_h==1,0],x[y_h==1,1],color='yellow',label='cluster2',s=30)
plt.scatter(x[y_h==2,0],x[y_h==2,1],color='orange',label='cluster3',s=30)
plt.scatter(x[y_h==3,0],x[y_h==3,1],color='green',label='cluster4',s=30)
plt.scatter(x[y_h==4,0],x[y_h==4,1],color='pink',label='cluster5',s=30)
plt.grid(False)
plt.title('Clusters of customers')
plt.xlabel('Annual_Income_(k$)')
plt.ylabel('Spending_Score')
plt.show()
```
